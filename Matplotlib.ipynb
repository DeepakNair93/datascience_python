# MATPLOTLIB - THE VISUALIZATION TOOL IN PYTHON

<b>Matplotlib is a library in Python which allows for the creation of interactive visualizations using the data we have in hand.

```python
#Let's start by importing the pre-requisite libraries

import matplotlib.pyplot as plt
import pandas as pd  
%matplotlib inline
```

```python
#Creating a list of numbers

l = [5,10,30,15,18,27,30,19,25,33,65,5]

#series to input to the plot
s = pd.Series(l, name='Readings')
```

```python
s.plot()
plt.show()
#Creating a plot object
```

<b> This is how simple it could be to create a plot using pandas datatypes & matplotlib
