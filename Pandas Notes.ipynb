What is a Series?

-- > One-dimensional array with axis labels (including time series).



```python
#Creating a series data type using Pandas.
a = pd.Series([1,-2,3], name="ID")
```

<b> Operations on Series Data Type.

```python
# Addition / Subtraction

print("Addition", a + 1000)
```

```python
#div / multiplication

print("Multiplication:", a *5)

print("\n Division", a/2)
```

```python
#Square using numpy
import numpy as np

print("Square:", np.square(a))

print("\n Mean of all elements", np.mean(a))
```

## DATAFRAMES
    
 DataFrame is another Datatype in Pandas. They are in the form of rows & columns. Just like a excel sheet table.
    
 Let's create a DataFrame.

```python
#creating 2 named series named s1 & s2.

s1 = pd.Series([100,250,370], index=['Fruits', 'Vegetables', 'Cereals'])

s2 = pd.Series([50,95,125], index =['Fruits','Junk-Food', 'Superfood'])

#creating dataframe named df using series s1 & s2
df = pd.DataFrame([s1,s2])
```

```python
print('Series 1 : \n', s1)

print('\n')

print('Series 2 : \n', s2)

print('\n')

print('DataFrame: \n', df)
print('\nData type:',type(df))
```

<b> In the above example we see that the indices of the series s1 & s2 are converted to Columns in DataFrame & the values are represented in separate rows with index 0 & 1. Where the values are not found, it imputs 'NaN' (Not a number) value in it.
